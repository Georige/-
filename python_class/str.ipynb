# 二进制算法

```python

x = float(input("Enter an decimal: "))
p = 0

while ((2**p)*x) % 1 != 0:
    print(f'Remainder = {(2**p)*x - int((2**p)*x)}')
    p += 1
num = int((2**p)*x)

if num < 0 :
    is_neg = True
    num = -num
else:
    is_neg = False

result = ''
if num == 0:
    result = '0'
while num > 0:
    result = str(num % 2) + result
    num = num // 2


for i in range(p - len(result)):
    result = "0" + result


result = result[0:-p] + "." + result[-p:]
if is_neg:
    result = "-" + result
print("The binary representation is", result)
```

```python
x = 54321
epsilon = 0.01
num_guesses = 0
guess = 0.0
incresement = 0.0001

while abs(guess**2 - x) >= epsilon and guess**2 <= x:
    guess += incresement
    num_guesses += 1
    if num_guesses % 100000 == 0:
        print('num_guesses = ', num_guesses, 'guess =', guess)
        print('difference = ', abs(guess**2 - x))


print('num_guesses = ', num_guesses)
print("last guess =", guess)
if abs(guess**2 - x) >= epsilon:
    print('Failed to find the square root of', x)
else:
    print(guess, 'is close to the square root of', x)
```

```python
x = 54321
epsilon = 0.0001
low = 0.0
high = x
guess = (high + low)/2
n = 0

while abs(guess**2 - x) >= epsilon:
    n += 1
    if guess**2 < x:
        low = guess
    else:
        high = guess
    guess = (high + low)/2
    
print('num of loops = ', n)
print(guess, 'is close to the square root of', x)
```
